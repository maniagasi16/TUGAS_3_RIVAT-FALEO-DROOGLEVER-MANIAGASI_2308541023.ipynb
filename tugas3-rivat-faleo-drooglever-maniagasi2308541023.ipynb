```python
umur = float(input("umur anda adalah"))
if umur >=60:
    print ("Anda sudah Lansia")
if umur <= 60 and umur >= 20:
    print("Anda sudah Dewasa")
if umur <=17 and umur >= 12:
    print("Anda masih Remaja")
if umur <=11 and umur >= 2:
    print("Anda masih Anak-anak")
if umur <= 1 and umur >= 0:
    print ("Anda masih Balita")
```

```python
total = 0
nilai = 1

while nilai <= 10:
    if nilai % 2 == 0:
        total += nilai
    nilai += 1

print("Total bilangan genap dari 1 hingga 10 adalah:", total)
```
